# Ingestão (Bronze)

Pra esse trabalho, fiz o upload de 2 arquivos da base [Yelp Dataset](https://www.kaggle.com/datasets/yelp-dataset/yelp-dataset/data): o `business.json` e `review.json`. 

Obs.: Devido à limitação de upload de arquivos da versão Community do Databricks (5 GB), foi utilizada uma amostra do conjunto de dados de avaliações (review.json), composta por 3 milhões de registros.

```python

#caminho + leitura do json

business_path = "dbfs:/Volumes/workspace/default/raw_yelp/yelp_academic_dataset_business.json"
review_path   = "dbfs:/Volumes/workspace/default/raw_yelp/yelp_review_sample.json"
df_business_raw = spark.read.json(business_path)
df_review_raw   = spark.read.json(review_path)

display(df_business_raw.limit(5))
display(df_review_raw.limit(5))
```

```python
#persite

(df_business_raw
 .write
 .mode("overwrite")
 .format("delta")
 .saveAsTable("bronze_yelp_business_raw"))
(df_review_raw
 .write
 .mode("overwrite")
 .format("delta")
 .saveAsTable("bronze_yelp_review_raw"))
```

```python
#validando se o persiste funcionou
print("business rows:", df_business_raw.count())
print("review rows:", df_review_raw.count())
```
